Выберем 5 различных аугментаций, пусть это будет:
1) Pad 
2) RandomAffine
3) ColorJitter
4) GaussianBlur
5) RandomHorizontalFlip

```python
import torch
import torchvision.transforms as tns
import matplotlib.pyplot as plt
torch.manual_seed(42)
img = torch.randint(low=0, high = 256, size = (256, 256))
plt.imshow(img)
```

```python
img = img.reshape((1, 1, 256, 256))
op = tns.Compose([
    tns.Pad(1),
    tns.RandomAffine(45.), 
    tns.ColorJitter(0.1, 0.5, 0.3, 0.2),
    tns.GaussianBlur(31),
    tns.RandomHorizontalFlip()
])

img = op(img)
```

```python
shape = img.shape
img = img.reshape((shape[2], shape[3]))
plt.imshow(img)
```
